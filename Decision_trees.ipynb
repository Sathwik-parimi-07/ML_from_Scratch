```python
def compute_entropy(y):
    entropy = 0.0
    if len(y) != 0:
        p1 = np.mean(y == 1)
        epsilon = 1e-9
        if p1 > 0 and p1 < 1:
            entropy = -p1 * np.log2(p1 + epsilon) - (1 - p1) * np.log2(1 - p1 + epsilon)
    return entropy
```

```python
def split_continuous_feature(X, y, feature, threshold):
    left_mask = X[:, feature] <= threshold
    right_mask = ~left_mask
    return X[left_mask], y[left_mask], X[right_mask], y[right_mask]
```

```python
def compute_information_gain_continuous(X, y, feature):
    best_gain = -1
    best_threshold = None
    unique_values = np.unique(X[:, feature])

    for threshold in unique_values:
        X_left, y_left, X_right, y_right = split_continuous_feature(X, y, feature, threshold)
        if len(y_left) == 0 or len(y_right) == 0:
            continue

        p_left = len(y_left) / len(y)
        p_right = len(y_right) / len(y)
        gain = compute_entropy(y) - (
            p_left * compute_entropy(y_left) + p_right * compute_entropy(y_right)
        )

        if gain > best_gain:
            best_gain = gain
            best_threshold = threshold

    return best_gain, best_threshold
```

```python
def find_best_split(X, y):
    best_gain = -1
    best_feature = None
    best_threshold = None

    for feature in range(X.shape[1]):
        gain, threshold = compute_information_gain_continuous(X, y, feature)
        if gain > best_gain:
            best_gain = gain
            best_feature = feature
            best_threshold = threshold

    return best_feature, best_threshold
```

```python
def build_tree(X, y, depth=0, max_depth=None):
    if len(set(y)) == 1 or (max_depth is not None and depth >= max_depth):
        return {"type": "leaf", "class": int(np.mean(y) > 0.5)}

    best_feature, best_threshold = find_best_split(X, y)
    if best_feature is None:
        return {"type": "leaf", "class": int(np.mean(y) > 0.5)}

    X_left, y_left, X_right, y_right = split_continuous_feature(X, y, best_feature, best_threshold)

    if len(y_left) == 0 or len(y_right) == 0:
        return {"type": "leaf", "class": int(np.mean(y) > 0.5)}

    left_subtree = build_tree(X_left, y_left, depth + 1, max_depth)
    right_subtree = build_tree(X_right, y_right, depth + 1, max_depth)

    return {
        "type": "node",
        "feature": best_feature,
        "threshold": best_threshold,
        "left": left_subtree,
        "right": right_subtree,
    }
```

```python
def predict_single(x, tree):
    if tree["type"] == "leaf":
        return tree["class"]
    if x[tree["feature"]] <= tree["threshold"]:
        return predict_single(x, tree["left"])
    else:
        return predict_single(x, tree["right"])
```

```python
def predict(X, tree):
    return np.array([predict_single(x, tree) for x in X])
```

```python
# Build the decision tree
tree = build_tree(x[:1000], y[:1000], max_depth=5)
```

```python
# Optional: Visualize the tree structure
def visualize_tree(tree, depth=0):
    if tree["type"] == "leaf":
        print("\t" * depth + f"Leaf: Class={tree['class']}")
    else:
        print("\t" * depth + f"Feature {tree['feature']} <= {tree['threshold']}:")
        visualize_tree(tree["left"], depth + 1)
        visualize_tree(tree["right"], depth + 1)

print("\nDecision Tree Structure:")
visualize_tree(tree)
```

```python
def calculate_f1_score(y_true, y_pred):

    y_true = np.asarray(y_true)
    y_pred = np.asarray(y_pred)

    # True Positives (TP): Predicted 1 and Actual 1
    tp = np.sum((y_pred == 1) & (y_true == 1))

    # False Positives (FP): Predicted 1 and Actual 0
    fp = np.sum((y_pred == 1) & (y_true == 0))

    # False Negatives (FN): Predicted 0 and Actual 1
    fn = np.sum((y_pred == 0) & (y_true == 1))

    precision = tp / (tp + fp) if (tp + fp) > 0 else 0
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0

    f1_score = (2 * precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    return f1_score,precision,recall
```
